```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load 
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
mouse_df = pd.DataFrame(mouse_data)
trial_df = pd.DataFrame(trial_data)
merge_data = mouse_df.merge(trial_df, on = "Mouse ID")

# Display the data table for preview
merge_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
merge_data1 = merge_data.drop(columns = ["Mouse ID", "Metastatic Sites"])
tumor_grp= merge_data1.groupby(["Drug", "Timepoint"]).mean()
#Convert to DataFrame
# Preview DataFrame
tumor_grp.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
merge_data2 = merge_data.drop(columns = ["Mouse ID", "Metastatic Sites"])
tumor_grp1 = merge_data2.groupby(["Drug", "Timepoint"]).sem()
# Convert to DataFrame
tumor_grp1.head()
# Preview DataFrame
tumor_pivot=tumor_grp1.pivot_table(values='Tumor Volume (mm3)', index=['Timepoint'], columns=['Drug'])
tumor_pivot
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_df = tumor_grp.pivot_table(values='Tumor Volume (mm3)', index=['Timepoint'], columns=['Drug'])
# Preview that Reformatting worked
pivot_df
```

```python
# Generate the Plot (with Error Bars)
y_error = tumor_pivot.values.tolist()
tumor_response = pivot_df.plot(kind = 'line', subplots = False, figsize = (10,6), title = "Tumor Response to Treatment", grid = True, legend = True, xlim = [0,45], ylim = [0,75], yerr = y_error)
tumor_response.set_xlabel("Days")
tumor_response.set_ylabel("Average Tumor Volume (mm3)")
# Save the Figure
plt.savefig("err_plot_1.png")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
merge_data3 = merge_data.drop(columns = ["Mouse ID", "Tumor Volume (mm3)"])
tumor_grp2 = merge_data3.groupby(["Drug", "Timepoint"]).mean()
# Convert to DataFrame
# Preview DataFrame
tumor_grp2.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
merge_data4 = merge_data.drop(columns = ["Mouse ID", "Tumor Volume (mm3)"])
tumor_grp3 = merge_data4.groupby(["Drug", "Timepoint"]).sem()
# Convert to DataFrame
tumor_grp3.head()
# Preview DataFrame
meta_pivot=tumor_grp3.pivot_table(values='Metastatic Sites', index=['Timepoint'], columns=['Drug'])
meta_pivot
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_df2 = tumor_grp3.pivot_table(values='Metastatic Sites', index=['Timepoint'], columns=['Drug'])
# Preview that Reformatting worked
pivot_df2
```

```python
# Generate the Plot (with Error Bars)
y_error1 = meta_pivot.values.tolist()
meta_spread = pivot_df2.plot(kind = 'line', subplots = False, figsize = (10,6), title = "Metastatic Spread During Treatment", grid = True, legend = True, xlim = [0,45], ylim = [0,0.5], yerr = y_error1)
meta_spread.set_xlabel("Days")
meta_spread.set_ylabel("Metastatic Sites")
# Save the Figure
plt.savefig("err_plot_2.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
merge_data5 = merge_data.drop(columns = ["Metastatic Sites", "Tumor Volume (mm3)"])
mouse_count = merge_data5.groupby(["Drug", "Timepoint"]).count()
# Convert to DataFrame

# Preview DataFrame
mouse_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_df3 = mouse_count.pivot_table(values='Mouse ID', index=['Timepoint'], columns=['Drug'])
# Preview that Reformatting worked
pivot_df3
```

```python
# Generate the Plot (Accounting for percentages)
survival = pivot_df3.plot(kind = 'line', subplots = False, figsize = (10,6), title = "Survival During Treatment", grid = True, legend = True, xlim = [0,45], ylim = [0,26])
survival.set_xlabel("Days")
survival.set_ylabel("Living Mice")
# Save the Figure
plt.savefig("err_plot_3.png")
# Show the Figure
plt.show()
```
