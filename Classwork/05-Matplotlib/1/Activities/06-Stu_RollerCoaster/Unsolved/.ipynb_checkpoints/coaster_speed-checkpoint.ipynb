```python
%matplotlib notebook

#Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Data
danger_drop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railspeed = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
x_axis = [0, 20, 40, 60, 80, 100, 120]
```

```python
#Add a semi-transparent horizontal line at y = 0
plt.hlines(0, 0, 10, alpha=0.25)
```

```python
# Use dots or other markers for your plots, and change their colors
plt.plot(x_axis, danger_drop, color="red", label="Danger Drop")
plt.plot(x_axis, railspeed, color="red", label="RailSpeed")
```

```python
# Add labels to the x and y axes
plt.title("Coaster Speed Over Time")
plt.xlabel("Runtime (In Seconds)")
plt.ylabel("Coaster Speed (MPH)")
```

```python
#Format Table
# Set your x and y limits
plt.xlim(0, 120)
plt.ylim(0, 100)

# Set a grid on the plot
plt.grid()
```
