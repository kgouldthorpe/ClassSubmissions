```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
combined_df = pd.merge(dash_df, bitcoin_df, on="Date")
combined_df.head()
```

```python
# Rename columns so that they are differentiated
organized_df = combined_df.rename(columns={"Date": "Date", "Open_x": "Dash Open", "High_x": "Dash High", "Low_x": "Dash Low", "Close_x": "Dash Close", "Volume_x": "Dash Volume", "Market Cap_x": "Dash Market Cap"
                                 "Open_y": "Bitcoin Open", "High_y": "Bitcoin High", "Low_y": "Bitcoin Low", "Close_y": "Bitcoin Close", "Volume_y": "Bitcoin Volume", "Market Cap_y": "Bitcoin Market Cap"}
organized_df.head()
```
